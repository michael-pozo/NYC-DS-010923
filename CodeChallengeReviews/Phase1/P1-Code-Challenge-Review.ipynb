```python
# call_on_students(1)
```

Answer:

- We have two options for dealing with null values. Option 1 is to remove the rows with missing data from our dataset. This can be a good option when null values represent a very small percentage of our overall data. Removing values will prevent us from introducing bias into the dataset. Option 2 is to impute the null value based on the values near by it (frontfill/backfill) or replace the null value with a series descriptive statistic (mean/median/mode).
